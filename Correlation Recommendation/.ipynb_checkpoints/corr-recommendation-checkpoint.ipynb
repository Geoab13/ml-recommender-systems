```python
import pandas as pd
import numpy as np

user_ratings = pd.read_csv('../data/rating_final.csv')
cuisines = pd.read_csv('../data/chefmozcuisine.csv')
geodata = pd.read_csv("../data/geoplaces.csv", encoding='latin')
places = geodata[['placeID', 'name']]
```

```python
ratings = pd.DataFrame(user_ratings.groupby('placeID')['rating'].mean())
ratings['rating_count'] = pd.DataFrame(user_ratings.groupby('placeID')['rating'].count())
```

```python
ratings.sort_values('rating_count', ascending=False).head()
```

###### create a cross table of places and user ratings

```python
places_rating_crosstab = pd.pivot_table(data=user_ratings, values='rating', index='userID', columns='placeID')
places_rating_crosstab.head()
```

```python
tortas_ratings = places_rating_crosstab[135085]
tortas_ratings[tortas_ratings>=0]
```

```python
similar_to_tortas = places_rating_crosstab.corrwith(tortas_ratings)
```

```python
corr_tortas = pd.DataFrame(similar_to_tortas, columns=['PearsonR'])
#drop all null values
corr_tortas.dropna(inplace=True)
corr_tortas = corr_tortas.sort_values('PearsonR', ascending=False)
corr_tortas.head()
```

Filter out places that have less than 10 ratings since these results will not be significant.

```python
tortas_corr_summary = corr_tortas.join(ratings['rating_count'])
tortas_corr_summary[tortas_corr_summary['rating_count']>=10].head(10)
```

Print places that have the highest correlation with restaurant Tortas Locas Hipocampo (placeID = 135085).

```python
top_corr_places_tortas = pd.DataFrame([135076, 135085, 135066, 132754, 135045, 135062, 135028, 135042, 135046, 132872], index=np.arange(10), columns=['placeID'])

summary = pd.merge(top_corr_places_tortas, cuisines, on='placeID', how='left')
summary
```
